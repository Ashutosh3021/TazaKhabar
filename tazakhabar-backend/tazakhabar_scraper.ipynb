```python
# Generate market observation using LLM
from src.services.llm_service import summarize_top_news, generate_observation_text
from src.db.database import async_session
from sqlalchemy import select
from src.db.models import Trend, Observation
from datetime import datetime

# First, summarize unsummarized news
await summarize_top_news(top_n=20)
print("✓ News summarization complete")

# Get trending keywords
async with async_session() as session:
    stmt = select(Trend).order_by(Trend.percentage_change.desc()).limit(10)
    result = await session.execute(stmt)
    trends = result.scalars().all()
    
    booming = [t.keyword for t in trends if t.percentage_change > 0]
    declining = [t.keyword for t in trends if t.percentage_change < 0]
    
    print(f"Booming keywords: {booming}")
    print(f"Declining keywords: {declining}")

# Generate observation
if booming or declining:
    observation_text = await generate_observation_text(booming, declining)
    print(f"Observation: {observation_text[:200]}...")
    
    # Save observation
    async with async_session() as session:
        obs = Observation(
            week_start=datetime.utcnow(),
            text=observation_text,
        )
        session.add(obs)
        await session.commit()
    print("✓ Observation saved")
```

```python
# Final verification - show all table counts
conn = sqlite3.connect('tazakhabar.db')
cursor = conn.cursor()

cursor.execute("SELECT name FROM sqlite_master WHERE type='table'")
tables = cursor.fetchall()

print("=== Final Database Status ===")
for (table_name,) in tables:
    cursor.execute(f'SELECT COUNT(*) FROM {table_name}')
    count = cursor.fetchone()[0]
    print(f"{table_name}: {count} rows")

# Show sample data
print("\n=== Sample Jobs ===")
cursor.execute("SELECT title, company, location FROM jobs LIMIT 5")
for row in cursor.fetchall():
    print(f"  {row[0]} @ {row[1]} ({row[2]})")

print("\n=== Sample News ===")
cursor.execute("SELECT title, score, type FROM news LIMIT 5")
for row in cursor.fetchall():
    print(f"  {row[0][:50]}... (score: {row[1]}, type: {row[2]})")

print("\n=== Sample Trends ===")
cursor.execute("SELECT keyword, count, percentage_change FROM trends LIMIT 10")
for row in cursor.fetchall():
    print(f"  {row[0]}: {row[1]} ({row[2]:+.1f}%)")

conn.close()
```
